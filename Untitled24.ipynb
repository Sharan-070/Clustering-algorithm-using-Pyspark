```python
import pandas as pd
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
import matplotlib.pyplot as plt
from pyspark.ml.clustering import GaussianMixture
from pyspark.sql import SparkSession
from pyspark.ml.evaluation import ClusteringEvaluator

# Create SparkSession
spark = SparkSession \
        .builder \
        .master('spark://192.168.64.1:7077')\
        .appName("GMM")\
        .getOrCreate()

# Read CSV file into a PySpark dataframe
df = spark.read.csv('insurance.csv', header=True, inferSchema=True)

# Create a vector assembler to combine numerical columns into a features column
assembler = VectorAssembler(inputCols=["age", "expenses","bmi"], outputCol='features')
df = assembler.transform(df)

# Train the Gaussian Mixture Model
gmm = GaussianMixture().setK(4).setSeed(123)
model = gmm.fit(df)

# Make predictions on the dataframe
predictions = model.transform(df)

# Extract features into separate columns
pandas_df_pyspark = predictions.select('features', 'prediction').toPandas()
pandas_df_pyspark[['feature1', 'feature2', 'feature3']] = pandas_df_pyspark['features'].apply(lambda x: pd.Series(x.toArray()))

# Calculate the silhouette score
evaluator = ClusteringEvaluator()
silhouette_pyspark = evaluator.evaluate(predictions)

# Plot scatter plot
colors = ['red', 'blue', 'green', 'yellow']
plt.scatter(pandas_df_pyspark['feature1'], pandas_df_pyspark['feature2'], c=pandas_df_pyspark['prediction'].apply(lambda x: colors[x]))
plt.show()
# Print the silhouette score
print("Silhouette Score for GMM: ", silhouette_pyspark)
```

```python
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Read CSV file into a pandas dataframe
df = pd.read_csv('insurance.csv')

# Select numerical columns
df_num = df[['age', 'expenses', 'bmi']]

# Train the Gaussian Mixture Model
gmm = GaussianMixture(n_components=4, random_state=123)
model = gmm.fit(df_num)

# Make predictions on the dataframe
predictions = model.predict(df_num)

# Add the predictions to the dataframe
df_num['prediction'] = predictions

# Extract features into separate columns
pandas_df_sklearn = df_num.copy()

# Calculate the silhouette score
silhouette_sklearn = silhouette_score(df_num, predictions)

# Plot scatter plot
colors = ['red', 'blue', 'green', 'yellow']
plt.scatter(pandas_df_sklearn['age'], pandas_df_sklearn['expenses'], c=pandas_df_sklearn['prediction'].apply(lambda x: colors[x]))
plt.show()

# Print the silhouette score
print("Silhouette Score with scikit-learn: ", silhouette_sklearn)
```

```python
import pandas as pd
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from pyspark.sql import SparkSession
# Create SparkSession
spark = SparkSession \
        .builder \
        .master('spark://192.168.64.1:7077')\
        .appName("DBSCAN")\
        .getOrCreate()
# Read CSV file into a PySpark dataframe
df = spark.read.csv('insurance.csv', header=True, inferSchema=True)

# Create a vector assembler to combine numerical columns into a features column
assembler = VectorAssembler(inputCols=["age", "expenses","bmi"], outputCol='features')
df = assembler.transform(df)

# Extract features into separate columns
pandas_df_pyspark = df.select('features').toPandas()
pandas_df_pyspark[['feature1', 'feature2', 'feature3']] = pandas_df_pyspark['features'].apply(lambda x: pd.Series(x.toArray()))

# Use DBSCAN to cluster the data
dbscan = DBSCAN(eps=2, min_samples=5)
labels = dbscan.fit_predict(pandas_df_pyspark[['feature1', 'feature2', 'feature3']])

# Calculate the silhouette score
silhouette_sklearn = silhouette_score(pandas_df_pyspark[['feature1', 'feature2', 'feature3']], labels)
# Plot scatter plot
colors = ['red', 'blue','green']
plt.scatter(pandas_df_pyspark['feature1'], pandas_df_pyspark['feature2'], c=labels, cmap='viridis')
plt.show()
# Print the silhouette score
print("Silhouette Score dbscan: ", silhouette_sklearn)
```

```python
import pandas as pd
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession \
        .builder \
        .master('spark://192.168.64.1:7077')\
        .appName("Agglomerative Clustering")\
        .getOrCreate()

# Read CSV file into a PySpark dataframe
df = spark.read.csv('insurance.csv', header=True, inferSchema=True)

# Create a vector assembler to combine numerical columns into a features column
assembler = VectorAssembler(inputCols=["age", "expenses","bmi"], outputCol='features')
df = assembler.transform(df)

# Extract features into separate columns
pandas_df_pyspark = df.select('features').toPandas()
pandas_df_pyspark[['feature1', 'feature2', 'feature3']] = pandas_df_pyspark['features'].apply(lambda x: pd.Series(x.toArray()))

# Use Agglomerative Clustering to cluster the data
agg_clustering = AgglomerativeClustering(n_clusters=4)
labels = agg_clustering.fit_predict(pandas_df_pyspark[['feature1', 'feature2', 'feature3']])

# Calculate the silhouette score
silhouette_sklearn = silhouette_score(pandas_df_pyspark[['feature1', 'feature2', 'feature3']], labels)
# Plot scatter plot
colors = ['red', 'blue', 'green', 'yellow']
plt.scatter(pandas_df_pyspark['feature1'], pandas_df_pyspark['feature2'], c=labels, cmap='viridis')
plt.show()
# Print the silhouette score
print("Silhouette Score Agglomerative Clustering: ", silhouette_sklearn)
```

```python
import pandas as pd
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
import matplotlib.pyplot as plt
from sklearn.cluster import OPTICS
from sklearn.metrics import silhouette_score
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder.master('spark://192.168.64.1:7077').appName("OPTICS").getOrCreate()

# Read CSV file into a PySpark dataframe
df = spark.read.csv('insurance.csv', header=True, inferSchema=True)

# Create a vector assembler to combine numerical columns into a features column
assembler = VectorAssembler(inputCols=["age", "expenses","bmi"], outputCol='features')
df = assembler.transform(df)

# Extract features into separate columns
pandas_df_pyspark = df.select('features').toPandas()
pandas_df_pyspark[['feature1', 'feature2', 'feature3']] = pandas_df_pyspark['features'].apply(lambda x: pd.Series(x.toArray()))

# Use OPTICS clustering to cluster the data
optics = OPTICS(eps=2, min_samples=5)
labels = optics.fit_predict(pandas_df_pyspark[['feature1', 'feature2', 'feature3']])

# Calculate the silhouette score
silhouette_sklearn = silhouette_score(pandas_df_pyspark[['feature1', 'feature2', 'feature3']], labels)
# Plot scatter plot
colors = ['red', 'blue', 'green', 'yellow']
plt.scatter(pandas_df_pyspark['feature1'], pandas_df_pyspark['feature2'], c=labels, cmap='viridis')
plt.show()
# Print the silhouette score
print("Silhouette Score OPTICS: ", silhouette_sklearn)
```
